```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from tqdm import trange, tqdm
```

```python
X, y = load_diabetes(return_X_y=True)
X /= X.std(axis=0)  # Standardize data (easier to set the l1_ratio parameter)
```

```python
alphas = np.linspace(10**(-3), 2, 1000)
scores_MSE = []
wieghts = []

for an_alpha in tqdm(alphas):
    regression = linear_model.Lasso(alpha=an_alpha)
    regression.fit(X, y)
    scores_MSE.append(mean_squared_error(y, regression.predict(X)))
    wieghts.append(regression.coef_)

plt.figure(figsize=[16, 12])
for i in range(len(wieghts[1])):
    plt.plot(np.log(alphas), np.array(wieghts)[:, i], alpha=0.9)

plt.xlabel("log(alpha)", fontsize=20)
plt.ylabel("Weight", fontsize=20)
plt.show()
```
